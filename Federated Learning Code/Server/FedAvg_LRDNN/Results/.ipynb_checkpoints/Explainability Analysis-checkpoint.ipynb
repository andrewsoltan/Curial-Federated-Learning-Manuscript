```python
#Define Feature List
featureList = pd.Series(['ALBUMIN', 'ALK.PHOSPHATASE', 'ALT', 'BASOPHILS', 'BILIRUBIN', 'CREATININE', 'CRP', 'EOSINOPHILS', 'HAEMATOCRIT', 'HAEMOGLOBIN', 'LYMPHOCYTES', 'MEAN CELL VOL.', 'MONOCYTES', 'NEUTROPHILS', 'PLATELETS', 'POTASSIUM', 'SODIUM', 'UREA', 'WHITE CELLS', 'eGFR', 'Respiratory Rate', 'Heart Rate', 'Systolic Blood Pressure', 'Temperature Tympanic', 'Oxygen Saturation', 'Delivery device used', 'Diastolic Blood Pressure'])
```

```python
basepath = 'Results/'
```

```python
######## NB: PRIOR TO GENERATING CHARTS, ENSURE COMPLETED WITH FINAL SETTINGS
```

```python
#Load Final Global Models
with open(basepath+'SavedModel_LR Global Model.pkl', "rb") as input_file:
    LR_Global_Model = pickle.load(input_file)
with open(basepath+'SavedModel_DNN Global Model.pkl', "rb") as input_file:
    DNN_Global_Model = pickle.load(input_file)
```

```python
#Plot LR Global Model relative feature importances - coefficients as part of a total
shownumber = 20
w0 = LR_Global_Model.intercept_[0]
w = LR_Global_Model.coef_[0]
#Convert LR coefficients to scalars (remove vector/directionality), calculate proportion from total sum
feature_importance = pd.Series(abs(w)/sum(abs(w)), index=featureList).nlargest(shownumber).sort_values(ascending=False)
ax = feature_importance.plot.barh(x='feature', y='Relative Feature Importance')
plt.ylabel("Feature")
plt.xlabel("Relative Importance")
plt.show()
```

```python
#Re-plot LR feature importances using seaborn, for improved colour
ax = sns.barplot(feature_importance.values, feature_importance.index)
ax.set_xlabel('Relative Importance')
```

```python
#Save CSV of feature_impostances
feature_importance.to_csv("Logistic Regression - Global Model Relative Importances.csv")
```
